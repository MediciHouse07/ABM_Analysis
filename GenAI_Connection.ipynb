```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

# Load the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')


def get_completion_from_messages(messages, model=model, temperature=0.7, max_tokens=100):
    # Convert messages (a list of dicts) into a single prompt string
    prompt = "\n".join([msg["content"] for msg in messages])

    # Tokenize input
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate output
    output = model.generate(input_ids,
                            max_length=len(input_ids[0]) + max_tokens,
                            temperature=temperature,
                            pad_token_id=tokenizer.eos_token_id)

    # Decode and return the generated text
    response_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # Return only the newly generated part
    return response_text[len(prompt):].strip()

# Example usage
# messages = [{"role": "user", "content": "What is the capital of France?"}]
print(get_completion_from_messages(messages))
```
